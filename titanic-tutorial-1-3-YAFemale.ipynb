```python
df = pd.DataFrame(train_data)
# assume 'df' is your DataFrame and 'column_x' is the column of interest
null_rows = df[df['Age'].isnull()]

print(null_rows)
```

None of the Age columns have a corresponding Cabin column.  4 females w/o Ages.

```python
# replacing null values in Age with Mode
#inplace error
#train_data['Age'].fillna(train_data['Age'].mode().iloc[0], inplace = True)
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].mode().iloc[0])

train_data.info()
```

Looking at data Fare, sex, and age are factor for survival.  Young females that paid higher Fare had a higher chance to survive.  Majority of the passengers did not have a Parent or sibling and when did it had no impact.

```python
#df = pd.DataFrame(train_data)
mode_imputation = df['Age'].fillna(df['Age'].mode().iloc[0])
print("\nImputation using Mode:")
print(mode_imputation)
mode_imputation.info()

median_imputation = df['Fare'].fillna(df['Fare'].median())
print("\nImputation using Median:")
print(median_imputation)
```

```python
from sklearn.ensemble import RandomForestClassifier

#y = train_data["Survived"]
y = df["Survived"]
print(y)
```

Update pClass with a fare class value instead.

```python
conditions = [
    df['Fare'] <= 7.91,
    (df['Fare'] > 7.91) & (df['Fare'] <= 31.00),
    df['Fare'] > 31.00
]
values = [1, 2, 3]

df['Fclass'] = np.select(conditions, values, default=0)
print(df[['PassengerId', 'Fare', 'Fclass']])
```

```python
#add Fclass to test data too.
conditions = [
    test_data['Fare'] <= 7.91,
    (test_data['Fare'] > 7.91) & (test_data['Fare'] <= 31.00),
    test_data['Fare'] > 31.00
]
values = [1, 2, 3]

test_data['Fclass'] = np.select(conditions, values, default=0)
```

Case sensitive 'PClass' <> 'Pclass'

```python
features = ["Fclass", "Sex", "Age"]
#X = pd.get_dummies(train_data[features])
X = pd.get_dummies(df[features])
print(X)
```

```python
X_test = pd.get_dummies(test_data[features])

# model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
# model.fit(X, y)
# predictions = model.predict(X_test)

# Automatically address null values
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# valid options are 'mean', 'median', 'most_frequent', or 'constant'
imputer = SimpleImputer(strategy='most_frequent')
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)

pipeline = Pipeline(steps=[('imputer', imputer), ('model', model)])
pipeline.fit(X, y)
predictions = pipeline.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})

output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
